#### Merge in Pandas
    -> jaise sql me tables ko join krte the same
    -> bas yha hme direct excel/csv file available h
    -> inner, left, right, outer merge
    -> NaN is null
    -> df1 --> left table, df2--> right table

    #Concat(for union 2 tables)
     ->stack/ join/ append one data frame on another
     ->to execute concat ensure column name should be common (both table me same column exit must)
     ->if objective is to combnine monthly data to master table data



```python
df1 = pd.DataFrame({'id':[1,2,3,4,5], 'Maths':[87,98,45,67,59] })
df2 = pd.DataFrame({'id':[3,4,5,6,7], 'Physics':[89,85,86,45,90]})

print(df1.merge(df2, on ='id', how ='inner'))#common in both
print(df1.merge(df2, on = 'id', how = 'left'))#left common data
print(df1.merge(df2, on = 'id', how = 'right'))#right common data
print(df1.merge(df2, on = 'id', how = 'outer')) # all data show common + incommon
print(pd.concat([df1,df2]))# union of both table, common cheze repeat hongi
```

```python
df3 = pd.DataFrame({'id':[1,2,3,4,5], 'Maths':[87,98,45,67,59], 'School':['A','B','C','B','A']})
df4 = pd.DataFrame({'id':[3,4,5,6,7], 'Physics':[89,85,86,45,90], 'School':['C','B','C','B','A']})
df3.merge(df4, on = ['id','School'], how = 'inner')
```

#### HOW to HANDLE missing values
     -> dropna: to remove the null value of a data frame 
        -> df1.dropna()
        -> to make save update simple re-assign to original column

           #example -> df1 = df1.dropna()
                        or
                       df1.dropna(inplace = True)   Best workflow (to avoid this mistake in future) because it removes all NaN places 

    -> fillna: to replace that missing value with something else if inplace = true
        -> To replace your missing values --> mean, median, mode --> these options are best
        -> replacing missing values with average(mean) --> one of the best option # stats property
        -> agr pemanent change krna to fir dubara df1 me hi assign ke dena

           #example -> df1['col'] = df1['col'].fillna(value)
                        or
                       df1.fillna({'col name': value}, inplace=True)


      NOTE - ****** 
          If you drop NaN → the row is deleted forever and then you try fill these places nothing happen because places removed
          If you fill NaN → the row is kept and value is replaced means direclty fill the values at NaN you want
