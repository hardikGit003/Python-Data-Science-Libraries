# Statistics

It is a branch of mathemathics that deals with collection of data , presentation of data , Analysis of the data to extract some meaningful information .

It is divided into 2 parts -

1) **Descriptive statistics** - ( Mathematical operations )

It is a branch of statistics that involves collection and interpretation of data .

Descriptive stats is also divided into 2 parts -

-> **Measure of Central Tendancy**  : It is used to indicate where does the centre of my data lies .

Mean

Median

Mode


-> **Measure of Dispersion** : It is used to indicate how much my data is spread out / Dispersed in all the direction .

Range

Variance

Standard Deviation

Percentile

Quartile

Outliers


--------------------------------------------------------------------------


2) **Inferential statistics** - ( Prediction and conclusion )

It is a branch of statistics that deals with making prediction and  drawing conclusion based on sample data collected from population .

Inferential stats is also divided into 2 parts .


 **Hypothesis Testing**

-> Null hypothesis

-> Alternate hypothesis

#### Measure of Dispersion

```python

#Variance - the distrubution of the data, How the data points are present wrt the data set
#[5.9, 5.8 ,5.6 ,5.7,6.1]  - heights   5.8   - low variance

#[20k, 300k, 3000k, 2k, 100rs] - salary  600k  - High Variance
"""Low Variance
Eg: [ 12 ,13 ,10 , 9 ,19 ,16 ,15]
In the above list all the data points are closer to the mean of the data


High Variance
Eg: [ 100 , 99999 , 32453675 , 1 ,4 ,67 ,345645745 ]  
In the above list all the data points are far away from the mean of the data
"""



```

```python
l= [ 12 ,13 ,10 , 9 ,19 ,16 ,15]
print(stats.variance(l)) # Result low variance
print(stats.mean(l)) # for verification
```

```python
x=[ 100 , 99999 , 32453675 , 1 ,4 ,67 ,345645745 ]
print(stats.variance(x))  # Result high variance
print(stats.mean(x)) # Can see high mean 
```

```python
# Standard Deviation - The Magnitude by which the data points are separated wrt mean
#                    - or SD = square root of varience

l=[ 12 ,13 ,10 , 9 ,19 ,16 ,15]
print(stats.mean(l))
print(stats.stdev(l))
```

Types of Sampling

1) Random Sampling:

Definition: Every member of the population has an equal chance of being selected.

Example: Drawing names out of a hat to select participants for a survey.

2) Systematic Sampling:

Definition: A sample is chosen by selecting every 'n-th' member of the population after a random starting point.

Example: Selecting every 10th person on a list to participate in a study.

3) Stratified Sampling:

Definition: The population is divided into distinct subgroups (strata), and a random sample is taken from each subgroup.

Example: Dividing a population by gender and then randomly selecting equal numbers from each group.

4) Cluster Sampling:

Definition: The population is divided into clusters, usually based on geography or other natural groupings, and then a whole cluster is randomly selected for the sample.

Example: Selecting random schools in a district and surveying all students in those schools.
