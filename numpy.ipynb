#### Function define Shape & Size of Array
    # np.zeros([3,2])  or np.zeros([m,n])
      -> creates array of size 3X2 filled with float 0. 

    # np.random.random([3,2]) or np.random.random([m,n])
      -> creates array of size 3X2 having random number b/w 0 to 1
    
    # np.random.randint(1,10) or (m,n) range can be any
     -> generate single random value b/w 1 and 10

    # np.random.randomint(1,10,[3,4]) or (range,[m,n])
     -> generate array of size 3X4 having random no. b/w 1 to 10
    
    # np.full([m,n],any data)
     -> return array of size m,n full with data given

    # np.arange(1,5) or np.arange(m,n->exclude)
     -> return array [1,2,3,4]
    
    # np. arange(0,10,2) or np.arange(start,end,skip)
     -> retrun array by skipping the given size     
    
    # np.linspace(0,10,6) or np.linspace(range,parts)
     -> return array of 6 number b/w 1 to 10 haivng equal differece

```python
ar1 = np.zeros([3,2]) 
print(ar1)

ar2 = np.random.random([3,2])
print(ar2)
      
ar3 = np.random.randint(1,10)
print(ar3)
     
ar4 = np.random.randint(1,10,[3,4])
print(ar4)
     
ar5 =  np.full([3,5],7)
print(ar5)
ar6 =  np.full([3,5],'h')
print(ar6)
   
ar7 =  np.arange(1,5)
print(ar7)
    
ar8 =  np. arange(0,10,2)
print(ar8)
       
ar9 =  np.linspace(0,10,6)
print(ar9)
    
```

#### Reshape,Changing dimensions of Array
     # np.reshape(dimesnions)  where dimensions can be any = 2D,3D,4D............nD
      -> note product of dimension you creating is equal to size of array

     # Dot operator for execute many funtions
      -> np.arange(1,25).reshape(8,4)
      -> np.arange(1,13).reshape(3,4).dtype

```python
# suppose arr is array of size 24 elements
arr = np.arange(1,25)
print(arr)
print(arr.size)

ar1 = arr.reshape(6,4)  # size 6X4 = 24
print(ar1)

ar2 = arr.reshape(3,4,2) # size 3X4X2 = 24
print(ar2)

# Dot operator for execute many funtions
ar3 = np.arange(1,25).reshape(8,3)
print(ar3)

# note we are assign the result to a new array
```

#### Array Mathematics
    # Sum function only work with array and its axis
      -> Columns Sum axis - 0
      -> Rows sum axis - 1
    
    # Any other function of mathematics that fall under numpy only work with number 
      -> add, subtract, divide, multiply, log, exp, sqrt, sin, cos, tan

    # Function work on a paticlar array
      -> mean, max, min, var

```python
brr1 = np.array([[10,20],[30,40]])
```

```python
print(np.sum(brr1)) # total sum of all elements in a array
print(np.sum(brr1,axis = 0)) # sum column wise
print(np.sum(brr1,axis = 1)) # sum row wise

```

```python
print(np.mean(brr1))
print(np.min(brr1))
print(np.max(brr1))
print(np.var(brr1))
```

```python
print(np.add(10,20))
print(np.subtract(100,20))
print(np.multiply(10,20))
print(np.divide(10,20))
print(np.log(20))
print(np.exp(20))
print(np.sqrt(20))
print(np.sin(20))
print(np.cos(20))
```

#### Taking Subset From Array
     -> Using slicing concept 
      -> arrray[start:end,start:end] ->  [for rows,for cols] , note-> end value does'nt include by default
    
     -> if start value is not mentioned, python considers 0 by default
     -> if end value is not mentioned it extract till the last"
        

```python
brr2 = np.arange(1,25).reshape(6,4)
print(brr2)
```

```python
print(brr2[1:5,0:3])

print(brr2[2:6,0:4])
print(brr2[2:6,:4])
print(brr2[2:,:4])

# Q-> Extract from 2nd row till last row and all column

print(brr2[1:])

# Q-> Extract all rows and some columns
print(brr2[:,1:5])
```

#### Array Manipulation
    # Concatenate Vertically
     -> np.concatenate([ar1,ar2], axis = 0) 
     -> np.row_stack([ar1,ar2])
     -> np.vstack([ar1,ar2])

    # Concatenate Horizontally
     -> np.concatenate([ar1,ar2], axis = 1)
     -> np.column_stack([ar1,ar2])
     -> np.hstack([ar1,ar2])

```python
ar1 = np.array([[10,20],[30,40]])
ar2 = np.array([[50,60],[70,80]])

print(np.concatenate([ar1,ar2], axis = 0))
print(np.row_stack([ar1,ar2]))
print(np.vstack([ar1,ar2]))
```

```python
print(np.concatenate([ar1,ar2], axis = 1))
print(np.column_stack([ar1,ar2]))
print(np.hstack([ar1,ar2]))
```
