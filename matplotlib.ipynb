#### Scatter plot

#### Plotting Command or functions

    # Labeling x and y axis
     -> plt.xlabel('Naming') and plt.ylabel('Naming')

    # Title of Scatter plot
     -> plt.title('Title')

    # To enable grid
     ->plt.grid()

    # To show labels in scatter plot
     ->plt.legend()
     
    # Hide unwanted Message note -> this should be written in the last
     ->plt.show()


```python
# Sameple Data Set
A = [22,24,23,24,23]
B = [21,22,23,24,25]
C = [1,2,3,4,5]

plt.scatter(C,A)
plt.show()
```

```python
plt.scatter(C,A, marker = 'D', color = 'orange')
plt.xlabel('Months')
plt.ylabel('Product Sold')
plt.title('Sales performance of A')
plt.grid()
plt.show()
```

```python
plt.scatter(C,A, marker = 'D', color = 'orange', label = 'Person A')
plt.scatter(C,B, marker= 'X', color = 'green', label= 'Person B')
plt.xlabel('Months')
plt.ylabel('Product Sold')
plt.title('Sales performance of A')
plt.grid()
plt.legend() # to show lables in scatter you should use this command
plt.show()
```

#### To show 2 or more Scatter plot simultaneously same level
    -> Basically treating as a matrix part 
       (121) -> 1 row 2 column vali matrix me first place
       (122) -> 1 row 2 column vali matrix me second place
    -> plt.sublplot(matrix coulmn)
    -> plt.sublots_adjust(wspace = sapcing ) to give space between

```python
plt.subplot(121)
plt.scatter(C,A,label = 'Person A')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales performance of A')
plt.legend()


plt.subplot(122)
plt.scatter(C,B,label = 'Person B')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales performance of A')
plt.legend()

plt.subplots_adjust(wspace = 0.4)
plt.show()
```

#### Types of Plot
    -> Scatter we have read upper
      ->plt.scatter(x,y,marker,color,label)
      
    -> Line Plot
      ->plt.plot(x,y,label,color)

    -> Area Plot
      ->plt.stackplot(x,y,labels,color)

    #Note- To show lablels you should enable legend()

```python
plt.subplot(221)
plt.scatter(C,A,label = 'A') # scatter plot of C and A
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales performance of A')
plt.legend()

plt.subplot(222)
plt.plot(C,A,label = 'A',color='Red') #line plot of A and A
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales performance of A')
plt.legend()

plt.subplot(223)
plt.stackplot(C,A,labels='A',color='Green') #Area plot of A and A
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales performance of A')
plt.legend()


plt.subplots_adjust(wspace = 0.4, hspace = 0.5)
plt.show()
```

#### Histogram
     -> For continuous data to identify data distribution
     -> Bars are connected adjacent to each other

     -> plt.hist(data[col name], edgecolor='',facecolor='',bins =''), where edgecolor-color of edge, facecolor-color of bar

     ->Default bins and bucket is 10 when use histogram 
    
       

```python
data.head(2)
```

#As no. of unique values in continuous data is more random we cant create bar plot
#Histogram is another version of bar plot --> continuous data


#Why use Histogram
->So if we used bar plot, there will be many bars since there are many unique(discrete) values in diff data sets but for histogram it make bin and buckets for continous data(default 10) and calculate how many data poits falls in this range. Overall no of bars in histogram is less to a readable format.

```python
print(data['mpg'].nunique()) #no of unique values in mpg
print(data['mpg'].min()) #max of mpg
print(data['mpg'].max()) #min of mpg

print(data['weight'].nunique()) #number of unique values in weight
```

```python
#we will use mpg bcz some data is continuous in it
plt.hist(data['mpg'], edgecolor='Black', facecolor='Orange')
plt.show()
#we can see here are 10 bins(Data point 10,15,20------45) and how many numbers of bars are fall on bon range is frequency(10,20,30----75)
```

Note- Range of mpg value is 9.0 - 46.6 and default there are 10 buckets x-axis 10-45 so histogram calculating how many mpg's of paticular range is counted as a frequency at y-axis 0 - 75

```python
#Dry Verify the histogram 
#Divide the data into multiple buckets / bins--> 10 buckets by default
import numpy as np
np.linspace(9,46.6, 11)  #11 partition bcz 11 edges are preset

#As we can see the edges is similar to these 10 point values that is bins
```

#### Scatter Plot
