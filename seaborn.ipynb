```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_sample = pd.read_csv("AutoMy.csv")
data_sample.head()
```

```python
data_sample[['mpg','acceleration','weight','displacement']].corr()
```

Now Priting this correlation with more vizualiztion using seaborn heatmap

```python
sns.heatmap(data_sample[['mpg','acceleration','weight','displacement']].corr(), cmap = 'Blues', annot = True)
plt.savefig('Heatmap.png') # if no path is mentioned it will be saved in pwd (working directory)
```

```python
#we can see every plotting and we can correlate
sns.pairplot(data_sample[['mpg','acceleration','weight','displacement']])
plt.show()
```

```python
# similar as bar plot
sns.countplot(x = 'origin', data = data_sample)
plt.show()
```

```python
data_sample.groupby(['origin','cylinders']).size().reset_index()
```

```python
# Here we are using two coulmns three regions have different cylinders with frequency we can see in upper calculation
# region 1 has -> 4,6,8   region 2 has -> 4,5,6  region 3 has -> 3,4,6 
# cylinders are represented with different hue colours
sns.countplot(x = 'origin',data = data_sample, hue = 'cylinders')
plt.show()
```
